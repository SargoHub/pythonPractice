```python
import pandas as pd
```

**PANDAS SERIES**

```python
# using the Series() function
l=[1,2,'hello',4.55,9]
a=pd.Series(l)    # creating a pandas series using the Series() function
a
```

```python
a[2]  # printing any value
```

```python
type(a[2])
```

```python
# we can give custom index values as well
l2=[1,2,3,4,5]
s=pd.Series(l2,index=[8,9,10,11,12])  # index values can be passed as the argument in the Series() function
s
```

**PANDAS DATAFRAME**

```python
# it is 2D data structure like a table with rows and columns
d={'Name':['Virat','Gayle','ABD'],'Team':['RCB','RCB','RCB'],'Runs':[4500,5600,6000]}  # this is Dict type data
d
```

```python
f=pd.DataFrame(d)  # converting the dict data to pandas DataFrame table structure
f
```

```python
f['Name']  # indexing
```

```python
f['Name'][1]
```
