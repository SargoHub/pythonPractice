**Data Visulization**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=np.array([2,5,1,6,9])
plt.plot(a,linestyle='dotted',marker='D')
plt.xlabel('index')
plt.ylabel('Data Points')
plt.title('line plot')
plt.show()
```

```python
#linegraph using the matplotlib

plt.plot(a,linestyle='dotted',marker='D',color='red')
plt.xlabel('index')
plt.ylabel('Data Points')
plt.title('line plot')
plt.grid(True)
plt.show()
```

```python
a=[1,2,3,5]
b=[80,4,50,16]
plt.plot(a,b)
```

```python
# lineplot using the seaborn library

a=[1,2,3,5]
b=[80,4,50,16]
sns.lineplot(x=a,y=b,marker='o',linestyle='dotted',color='green')
```

```python
# BAR GRAPH

speed=[10,20,30]
car=[100,150,220]
plt.bar(speed,car)
```

```python
speed=[10,20,30]
car=[100,150,220]
plt.bar(speed,car,width=3.5,color='red',edgecolor='yellow')
plt.xlabel('speed')
plt.ylabel('car speed')
plt.grid()
plt.title('speed vs car speed')
plt.show()
```

```python
sns.barplot(x=speed,y=car)
plt.show()
```

```python
# horizontal bar graph using matplotlib
plt.barh(speed,car)
```

```python
# SCATTERPLOT USING MATPLOLIB
r=[5,7,8,10,11,16,19,23]
m=[34,45,55,23,78,66,59,50]
```

```python
plt.scatter(m,r,color='green',marker='*')
plt.grid(True)
plt.show()
```

```python
# SCATTERPLOT USING THE SEABORN LIBRARY

sns.scatterplot(x=m,y=r,marker='*',color='black',s=150)
plt.show()
```

```python
# HISTOGRAM

l=[15,65,18,28,68,47,39,33,68,25,52,65]
plt.hist(l)
plt.show()
```

```python
l=[15,65,18,28,68,47,39,33,68,25,52,65]
plt.hist(l,bins=6,edgecolor='pink')
plt.show()
```

```python
plt.hist(l,bins=6,edgecolor='pink',color='lightblue')
#plt.figure(figsize=(20,5))
plt.xlabel='values'
plt.ylabel='count'
plt.title='histogram'
plt.grid(True)
plt.show()
```

```python
# BOXPLOT using matplotlib

p=[25,28,29,29,30,27,38,40,-70,170,190,26,30,18,19]

plt.boxplot(p)  # this will give us the outlayers datapoints
plt.grid()
plt.show()
```

```python
# PIE CHART

a=[40,25,30,16]
label=['RCB','CSK','RR','PBKS']
plt.pie(a,labels=label)
plt.show()
```

```python
a=[40,25,30,16]
label=['RCB','CSK','RR','PBKS']
plt.pie(a,labels=label,autopct='%0.2f%%')
plt.show()
```

```python
a=[40,25,30,16]
label=['RCB','CSK','RR','PBKS']
plt.pie(a,labels=label,autopct='%0.2f%%',explode=[0,0,0.2,0],shadow=True)
plt.show()
```
